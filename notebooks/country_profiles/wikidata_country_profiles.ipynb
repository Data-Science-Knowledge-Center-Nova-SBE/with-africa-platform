```python
from SPARQLWrapper import SPARQLWrapper, JSON
import pandas as pd
endpoint_url = "https://query.wikidata.org/sparql"
query = """
SELECT ?park ?parkLabel ?country ?countryLabel ?countryCode ?area ?coordinate ?inception WHERE {
  ?park wdt:P31 wd:Q46169;           # Instance of national park
        wdt:P17 ?country;            # Located in country
        wdt:P2046 ?area;             # Area of the park
        wdt:P625 ?coordinate;
        wdt:P571 ?inception.
  ?country wdt:P30 wd:Q15;           # Country is in Africa
           wdt:P298 ?countryCode.    # ISO 3166-1 alpha-3 code

  SERVICE wikibase:label { bd:serviceParam wikibase:language "[AUTO_LANGUAGE],en". }
}
"""
sparql = SPARQLWrapper(endpoint_url)
sparql.setQuery(query)
sparql.setReturnFormat(JSON)
results = sparql.query().convert()
# Convert the results into a Pandas dataframe.
cols = results['head']['vars']
out = []
for row in results['results']['bindings']:
    item = []
    for c in cols:
        item.append(row.get(c, {}).get('value'))
    out.append(item)

df = pd.DataFrame(out, columns=cols)
df.head()
```

```python
from SPARQLWrapper import SPARQLWrapper, JSON
import pandas as pd
endpoint_url = "https://query.wikidata.org/sparql"
query = """
SELECT ?country ?countryLabel ?currency ?currencyLabel ?region ?regionLabel ?officialLanguage ?officialLanguageLabel ?capital ?capitalLabel ?emoji ?iso3 ?population WHERE {
  ?country wdt:P31 wd:Q6256;           # Instance of country
           wdt:P30 wd:Q15.             # In the continent of Africa

  OPTIONAL { ?country wdt:P38 ?currency. }          # Currency
  OPTIONAL { ?country wdt:P361 ?region. }           # Part of (region)
  OPTIONAL { ?country wdt:P37 ?officialLanguage. }  # Official language
  OPTIONAL { ?country wdt:P36 ?capital. }           # Capital
  OPTIONAL { ?country wdt:P1813 ?shortName. }       # Short name
  OPTIONAL { ?country wdt:P487 ?emoji. }            # Emoji regional indicator sequence
  OPTIONAL { ?country wdt:P298 ?iso3. }            # Emoji regional indicator sequence
  OPTIONAL { ?country wdt:P1082 ?population. }            # Emoji regional indicator sequence

  SERVICE wikibase:label { bd:serviceParam wikibase:language "[AUTO_LANGUAGE],en". }
}

"""
sparql = SPARQLWrapper(endpoint_url)
sparql.setQuery(query)
sparql.setReturnFormat(JSON)
sparql.addCustomHttpHeader("User-Agent", "WAAgent")  # Add a user agent

results = sparql.query().convert()
# Convert the results into a Pandas dataframe.
cols = results['head']['vars']
out = []
for row in results['results']['bindings']:
    item = []
    for c in cols:
        item.append(row.get(c, {}).get('value'))
    out.append(item)

df_country_information = pd.DataFrame(out, columns=cols)
df_country_information.head()
```
